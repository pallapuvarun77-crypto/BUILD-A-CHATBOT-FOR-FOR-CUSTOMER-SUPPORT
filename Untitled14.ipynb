```python
# ==========================================================
# TASK 3: INTELLIGENT CUSTOMER SUPPORT CHATBOT
# NLP-BASED | SINGLE FILE | INTERNSHIP READY
# ==========================================================

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -------------------------------
# STEP 1: FAQ DATASET (INTENTS)
# -------------------------------

questions = [
    "hello",
    "hi",
    "how can you help me",
    "how to track my order",
    "where is my order",
    "order status",
    "how can I get a refund",
    "refund policy",
    "return my money",
    "customer support contact",
    "how to contact support",
    "help me",
    "bye",
    "thank you"
]

answers = [
    "Hello! How can I assist you today?",
    "Hi! How may I help you?",
    "I can help you with orders, refunds, and support.",
    "You can track your order using the tracking link sent to your email.",
    "Please provide your order ID to check the order status.",
    "Your order status is available in the Orders section.",
    "Refunds are processed within 5 to 7 business days.",
    "Our refund policy allows returns within 10 days of delivery.",
    "Your money will be refunded to the original payment method.",
    "You can contact customer support via email or chat.",
    "Our support team is available 24/7 to assist you.",
    "Sure! Please tell me your issue.",
    "Goodbye! Have a great day ðŸ˜Š",
    "You're welcome! Happy to help ðŸ˜Š"
]

# -------------------------------
# STEP 2: VECTORIZE QUESTIONS
# -------------------------------

vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)

# -------------------------------
# STEP 3: CHATBOT FUNCTION
# -------------------------------

def chatbot_response(user_input):
    user_vector = vectorizer.transform([user_input])
    similarity = cosine_similarity(user_vector, question_vectors)
    best_match = similarity.argmax()
    score = similarity[0][best_match]

    if score < 0.25:
        return "Sorry, I couldn't understand that. Please try asking differently."
    else:
        return answers[best_match]

# -------------------------------
# STEP 4: CHAT LOOP
# -------------------------------

print("ðŸ¤– AI Customer Support Chatbot")
print("Ask me about orders, refunds, or support.")
print("Type 'bye' to exit\n")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Bot: Goodbye! Have a nice day ðŸ˜Š")
        break

    reply = chatbot_response(user_input)
    print("Bot:", reply)

# -------------------------------
# END OF TASK 3
# ==========================================================
```
